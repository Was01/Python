```python
dados=pd.read_csv('/content/Titanic-Dataset.csv')
dados.head()
```

```python
dados.isnull().sum()
```

```python
dados.shape
```

```python
dados.drop('Cabin',axis=1,inplace=True)
dados.head()
```

```python
dados.isnull().sum()
```

```python
dados['Embarked'].value_counts()
```

```python
dados['Embarked']=dados['Embarked'].fillna('S')
dados.isnull().sum()
```

```python
media_idade=  dados['Age'].mean()
dados['Age']=dados['Age'].fillna(media_idade)
dados.isnull().sum()
```
