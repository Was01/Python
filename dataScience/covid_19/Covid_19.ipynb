```python
import numpy as np
from urllib.request import urlretrieve
import pandas as pd
import matplotlib.pyplot as plt
```

## Casos confirmados de covid-19

```python
url='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/refs/heads/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'

urlretrieve(url,'covid_confirmados.csv')
```

```python
df_covid=pd.read_csv('/content/covid_confirmados.csv')
df_covid.head()
```

```python
df_covid.drop(['Lat','Long'],axis=1,inplace=True)  #Exluindo as colunas Lat e Long
df_covid.head()
```

```python
df_country=df_covid.groupby('Country/Region').sum()
```

```python
df_country.loc['Brazil']
```

```python
date=df_country.loc['Brazil'].index

date
```

```python
deaths=df_country.loc['Brazil'].values

deaths
```

```python
S_brazil=df_country.loc['Brazil']
S_brazil=S_brazil[S_brazil>0]
```

```python
tam=len(S_brazil)

plt.figure(figsize=(20,10))
plt.bar(S_brazil.index[tam-50:tam],S_brazil.values[tam-50:tam])
plt.xticks(rotation=90)
plt.show()
```
