```python
import pandas as pd
import plotly.express as px
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
```

```python
df.head()
```

## Histogramas

```python
px.histogram(df,x='total_bill',nbins=10,histnorm='probability density',marginal='box',title='Taxa paga no restaurante')
```
