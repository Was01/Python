```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,10,0.01)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label='sin(x)')
plt.plot(x,y2,label='cos(x)')
plt.grid(True)
plt.legend()
plt.xticks(np.arange(0,11))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Subplots

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(x,y1,label='sin(x)')
plt.subplot(1,2,2)
plt.plot(x,y2,label='cos(x)')
plt.show()
plt.tight_layout()
```
