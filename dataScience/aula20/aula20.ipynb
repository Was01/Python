```python
import pandas as pd
import numpy as np
import plotly.express as px

df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
df.head()
```

```python
df_agrupado_dia=df.groupby(['day'],sort=False).agg({'total_bill':'sum'}).reset_index()
df_agrupado_dia
```

```python
## Gr√°fico de Barros

px.bar(df_agrupado_dia,x='day',y='total_bill')
```

```python
df_agrupado_dia_sexo=df.groupby(['day','sex'],sort=False).agg({'total_bill':'sum'}).reset_index()
df_agrupado_dia_sexo
```

```python
px.bar(df_agrupado_dia_sexo,x='day',y='total_bill',color='sex',barmode='group')
```

```python
## Gr√°fico de dispers√£o
px.scatter(df,x='total_bill',y='tip',color='sex',hover_name='time')
```

```python
## Gr√°fico de setores
quantidade_pedidos_dia=df['day'].value_counts()

px.pie(quantidade_pedidos_dia,names=quantidade_pedidos_dia.index,values=quantidade_pedidos_dia.values)
```
