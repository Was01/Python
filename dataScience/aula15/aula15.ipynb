```python
import os # para executar comandos de diretório
from google.colab import drive # habilitar google drive no colab
drive.mount('/content/drive')
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/dados/Titanic-Dataset.csv')
df.head()
```

```python
def changeSurvided(elemento):
  if elemento==1:
    return 'Yes'
  else:
    return 'No'

df['Survived']=df['Survived'].apply(changeSurvided)
df.head()
```

```python
def changeFare(elemento):
  return f'${elemento}'

df['Fare']=df['Fare'].apply(changeFare)
df.head()
```

## Ordenação

```python
media_idade=df['Age'].mean()
df['Age']=df['Age'].fillna(media_idade)
df.isnull().sum()
```
