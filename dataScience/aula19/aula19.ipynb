```python
df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv')
df.head()
```

```python
df.tail()
```

```python
df_agrupado_mes=df.groupby('month',sort=False).mean().reset_index()
df_agrupado_mes.head()
```

```python
px.line(df_agrupado_mes,x='month',y='passengers')
```

```python
px.line(df,x='month',y='passengers',color='year')
```
