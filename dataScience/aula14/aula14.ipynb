## 14 - Junção de DataFrames

```python
import numpy as np
import pandas as pd
```

```python
dados=np.arange(1,10).reshape(3,3)
dados
```

```python
df1=pd.DataFrame(dados,columns=['A','B','C'])
df2=pd.DataFrame(dados,columns=['A','B','C'])
df3=pd.DataFrame(dados,columns=['D','E','F'])
```

```python
df1
```

```python
df2
```

```python
df3
```

## Concat



```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df3],axis=1)
```

## Merge

```python
df1=pd.DataFrame(dados,columns='A B C'.split())
df2=pd.DataFrame(dados,columns='D E C'.split())
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='C')
```

## Join

```python
df1=pd.DataFrame(dados,columns='A B C'.split(),index=[0,1,2])
df2=pd.DataFrame(dados,columns='D E F'.split(),index=[1,2,3])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how='right')
```

```python
df1.join(df2,how='outer')
```
