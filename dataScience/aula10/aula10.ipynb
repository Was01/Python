## Filtragem de dados

```python
df['Age']>25 #permite a visualização de valores booleanos de uma coluna
```

```python
df[df['Age']>25] #permite a visualização de linhas que atendem a uma condição
```

```python
df[df['Age']>df['Age'].mean()] #permite a visualização de linhas que atendem a uma condição
```
