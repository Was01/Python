```python
import pandas as pd
import numpy as np
```

```python
massa_corporal=[72,80,60,90,100]
altura=[180,170,175,174,185]
idade=[21,21,16,28,23]
covid_19=['não','não','não','sim','sim']

dados_covid=pd.DataFrame({'massa corporal(kg)':massa_corporal,'Altura(cm)':altura,'idade':idade,'covid-19':covid_19})

dados_covid.head()
```

```python
dados_covid.shape
```

```python
dados_covid_agrupado=dados_covid.groupby('covid-19').mean()
dados_covid_agrupado
```
